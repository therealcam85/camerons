```python
import yfinance as yf
import pandas as pd

def get_tesla_revenue():
    # Fetch Tesla's financial data
    tsla = yf.Ticker("TSLA")

    # Get financials (Income Statement)
    income_stmt = tsla.financials

    # Extract revenue row
    if "Total Revenue" in income_stmt.index:
        revenue_data = income_stmt.loc["Total Revenue"]
        df = pd.DataFrame({"Year": revenue_data.index.year, "Revenue (USD Millions)": revenue_data.values / 1e6})
        return df
    else:
        print("Revenue data not found.")
        return None

# Fetch and display Tesla revenue data
tesla_revenue_df = get_tesla_revenue()
if tesla_revenue_df is not None:
    # Use pandas display function instead of ace_tools
    #from IPython.display import display # If not already imported
    display(tesla_revenue_df) # This will display the DataFrame in the output
```
